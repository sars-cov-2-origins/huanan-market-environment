## January 1 and January 12 trips

```python
## First sampling trip: samples, stalls, positives, sequenced positives

jan1 = d.query("`Sampling.date` == '2020-01-01'")
print("{} Total January 1 samples".format(jan1.shape[0]))
jan1_pos = jan1.query("`SARS.CoV.2.qPCR.result` == 'Positive'")
print("{} Total January 1 Positives".format(jan1_pos.shape[0]))
print("{} Total January 1 stalls".format(jan1.Stall_corrected.unique().shape[0]))
print("{} Total January 1 merged stalls".format(jan1.Stall_corrected_merged.unique().shape[0]))

## Jan 1 samples which were sequenced
print("{} total January 1 samples sequenced".format(jan1[~jan1.Sequencing_Run.isna()].shape[0]))
```

```python
## Second sampling trip: samples, stalls, and sequenced negatives / positives

jan12 = d.query("`Sampling.date` == '2020-01-12'")
print("{} Total January 12 samples".format(jan12.shape[0]))
jan12_pos = jan12.query("`SARS.CoV.2.qPCR.result` == 'Positive'")
print("{} Total January 12 Positives".format(jan12_pos.shape[0]))
print("{} Total January 12 stalls".format(jan12.Stall_corrected.unique().shape[0]))
print("{} Total January 12 merged stalls".format(jan12.Stall_corrected_merged.unique().shape[0]))

## Jan 12 samples which were sequenced
print("{} total January 1 samples sequenced".format(jan12[~jan12.Sequencing_Run.isna()].shape[0]))
```

```python
## Jan 12 positive samples 
jan12_pos
```
